```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('usdeur.csv')
```

```python
print(df.to_string()) 
```

```python
mean = df['We buy'].mean()
print(mean)
print(1/mean)
```

```python
df['usd->eur'] = df['We buy'].apply(lambda x: 1/x)
```

```python
print(df.to_string())
```

```python
df.plot(y='usd->eur', use_index = True)
```

```python
df['diff']=df['usd->eur'].diff()
```

```python
print(df.to_string())
```

```python
df.plot(y='diff',use_index=True)
```

```python
df['ma'] = df['usd->eur'].rolling(window=5).mean()
df.plot(y = 'ma', use_index=True)
```

```python
df['diff_ma'] = df['diff'].rolling(window = 10).mean()
df.plot(y='diff_ma', use_index=True)
```

```python
df.plot(y=['usd->eur','ma'])
```
